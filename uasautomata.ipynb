```python
class FSMTopup:
    def __init__(self):
        self.transitions = {
            'q0': {'TOPUP': 'q1'},
            'q1': {
                '10000': 'q2',
                '20000': 'q3',
                '50000': 'q4'
            }
        }

        self.final_states = {
            'q2': ('Pulsa Rp10.000 berhasil, bonus 1GB aktif.', 10000, 1),
            'q3': ('Pulsa Rp20.000 berhasil, bonus 3GB aktif.', 20000, 3),
            'q4': ('Pulsa Rp50.000 berhasil, bonus 10GB aktif.', 50000, 10)
        }

        self.total_pulsa = 0
        self.total_kuota = 0

    def process(self, message):
        msg = message.strip().upper()

        if msg == "CEK PULSA":
            return f"Total pulsa Anda: Rp{self.total_pulsa:,}".replace(",", ".")  # Format pakai titik

        if msg == "CEK KUOTA":
            return f"Total kuota Anda: {self.total_kuota}GB"

        parts = msg.split()

        if len(parts) != 2:
            return "Mohon gunakan format: TOPUP (10000/20000/50000), atau CEK PULSA/KUOTA."

        command, amount = parts

        if command not in self.transitions['q0']:
            return "Format salah. Gunakan: TOPUP (10000/20000/50000), atau CEK PULSA/KUOTA."

        state1 = self.transitions['q0'][command]

        if amount not in self.transitions.get(state1, {}):
            return "Nominal salah. Gunakan 10000, 20000, atau 50000."

        final_state = self.transitions[state1][amount]
        response, pulsa, kuota = self.final_states.get(final_state, ("Transaksi gagal.", 0, 0))

        self.total_pulsa += pulsa
        self.total_kuota += kuota

        return response


fsm = FSMTopup()

while True:
    print("\nKetik SMS Anda:")
    print("- TOPUP <nominal> (contoh: TOPUP 10000)")
    print("- CEK PULSA")
    print("- CEK KUOTA")
    print("- KELUAR untuk keluar")

    user_input = input("SMS: ")

    if user_input.strip().upper() == "KELUAR":
        print("Terima kasih. Program selesai.")
        break

    print(fsm.process(user_input))
```
